````python
from graphviz import Digraph

def draw_file_system_tree():
    dot = Digraph()

    # Root
    dot.node('root', '/', shape='ellipse')

    # Directories
    dot.node('home', 'home', shape='ellipse')
    dot.node('var', 'var', shape='ellipse')
    dot.node('usr', 'usr', shape='ellipse')
    dot.node('pjd', 'pjd', shape='ellipse')
    dot.node('tmp', 'tmp', shape='ellipse')

    # Files
    dot.node('file1', 'file1', shape='rectangle')
    dot.node('file2', 'file2', shape='rectangle')
    dot.node('file3', 'file3', shape='rectangle')

    # Data
    dot.node('aaa', 'aaa', shape='circle')
    # dot.node('bbb', 'bbb', shape='circle')
    dot.node('ccc', 'ccc', shape='circle')
    # dot.node('ddd', 'ddd', shape='circle')
    dot.node('eee', 'eee', shape='circle')
    dot.node('fff', 'fff', shape='circle')

    # Connections
    dot.edges([('root', 'home'), ('root', 'var'), ('root', 'usr')])
    dot.edges([('home', 'pjd')])
    dot.edges([('var', 'file2'), ('var', 'tmp')])
    dot.edges([('usr', 'file1')])
    dot.edges([('pjd', 'file3')])

    dot.edges([('file3', 'fff')])
    dot.edges([('file2', 'eee')])
    dot.edges([('eee', 'ccc')])
    dot.edges([('file1', 'aaa')])


    # Set output format and render
    dot.format = 'svg'
    dot.render('file_system_tree', view=False)

    from IPython.display import SVG, display
    display(SVG(filename='file_system_tree.svg'))

draw_file_system_tree()

````
